# Running Instructions
This project requires the Rust tool set to build and run, get it at this link https://www.rust-lang.org/tools/install. Download rustup-init and run it, it should add the tools to your path automatically.

Navigate to the rust project with

`cd algorithms`

Build and run the rust project on the retail dataset with

`cargo run ../data/retail.dat`

You'll find the results in `algorithms/output`
